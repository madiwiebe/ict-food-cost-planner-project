```python
import pandas as pd
import numpy as np
```

```python
data_1 = pd.read_csv('farm_Sept2022_Sept2023.csv')
print("There are " + str(len(data_1)) + " observations in the retail prices dataset.")
print("There are " + str(len(data_1.columns)) + " variables in the dataset.")
```

```python
# drop Geography column
data_1.drop(['Geography'], axis=1, inplace=True)

# remove empty rows
data_1.dropna(subset=['Product Name'], inplace=True)

data_1.head(30)
```

```python
cols = data_1.columns.drop(['Product Name', 'Units'])
data_1[cols] = data_1[cols].apply(pd.to_numeric, errors='coerce')
data_1
```

```python
data_1 = data_1.fillna(0)
data_1
```

```python
data_1 = data_1.groupby(['Product Name','Units'], as_index=False).mean()
data_1
```

```python
data_1['Product Name'] = data_1['Product Name'].str.replace('\d+', '')
data_1
```
