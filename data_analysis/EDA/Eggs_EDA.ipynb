```python
import os
```

```python
# Specific Excel files
excel_files = ['Eggs2013.xlsx', 'Eggs2014.xlsx', 'Eggs2015.xlsx', 'Eggs2016.xlsx', 'Eggs2017.xlsx']
```

```python
combined_data_eggs = pd.DataFrame()
```

```python
# Loop through each Excel file and concatenate its data to the combined_data DataFrame
for file in excel_files:
    file_path = os.path.join('/Users/tativalentine/Desktop/FOOD/EDA', file)
    df = pd.read_excel(file_path)  
    combined_data_eggs = pd.concat([combined_data_eggs, df], ignore_index=True)
```

```python
print(combined_data_eggs.columns)
```

```python
# Line plot to visualize average prices over time
plt.figure(figsize=(12, 6))
for product in combined_data_eggs['Product Name'].unique():
    product_data = combined_data_eggs[combined_data_eggs['Product Name'] == product]
    plt.plot(product_data['Start Date'], product_data['Canada'], label=product)

plt.title('Average Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Average Price (Canada)')
plt.legend()
plt.show()
```

```python

# Feature engineering: Extract seasons (month-based seasons)
combined_data_eggs['Month'] = combined_data_eggs['Start Date'].dt.month
seasons = {
    1: 'Winter',
    2: 'Winter',
    3: 'Spring',
    4: 'Spring',
    5: 'Spring',
    6: 'Summer',
    7: 'Summer',
    8: 'Summer',
    9: 'Fall',
    10: 'Fall',
    11: 'Fall',
    12: 'Winter'
}
combined_data_eggs['Season'] = combined_data_eggs['Month'].map(seasons)

# Select relevant columns for provinces and Canada
provinces_columns = ['British Columbia', 'Alberta', 'Saskatchewan', 'Manitoba', 'Remaining Ontario',
                      'Ontario', 'Montreal', 'Remaining Quebec', 'Quebec', 'Maritimes', 'Newfoundland and Labrador', 'Canada']

# Melt the DataFrame to have 'Province' and 'Average Price' columns
melted_data = pd.melt(combined_data_eggs, id_vars=['Season', 'Product Name'], value_vars=provinces_columns,
                      var_name='Province', value_name='Average Price')
```

```python
import pandas as pd

# Assuming melted_data contains the 'Average Price' column
melted_data['Average Price'] = pd.to_numeric(melted_data['Average Price'], errors='coerce')

# Check the data types
print(melted_data.dtypes)

# Now, try the bar plot again
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 8))
sns.barplot(x='Season', y='Average Price', hue='Province', data=melted_data, palette='viridis')
plt.title('Average Prices of Provinces in Each Season')
plt.xlabel('Season')
plt.ylabel('Average Price')
plt.show()
```

```python
plt.figure(figsize=(16, 12))
sns.FacetGrid(melted_data, col='Province', col_wrap=3, height=4).map(sns.lineplot, 'Season', 'Average Price', ci=None, palette='viridis')
plt.subplots_adjust(top=0.9)
plt.suptitle('Trends of Average Prices for Each Province Across Seasons', fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(14, 8))
sns.boxplot(x='Season', y='Average Price', hue='Province', data=melted_data, palette='viridis')
plt.title('Boxplot of Prices for Provinces in Each Season')
plt.xlabel('Season')
plt.ylabel('Average Price')
plt.show()
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
# Select relevant columns
data_for_clustering = combined_data_eggs[['Start Date', 'Canada']]
```

```python
# Standardize the data
scaler = StandardScaler()
data_for_clustering['Canada_scaled'] = scaler.fit_transform(data_for_clustering[['Canada']])
```

```python
# Define the number of clusters
num_clusters = 3
```

```python
# Check and handle missing values
if data_for_clustering.isnull().values.any():
    # If there are missing values, you can drop them or impute them
    data_for_clustering = data_for_clustering.dropna()  # Example: Drop rows with missing values
    # Alternatively, you can impute missing values using methods like mean, median, or interpolation
```

```python
# Fit the k-means model
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
data_for_clustering['Cluster'] = kmeans.fit_predict(data_for_clustering[['Canada_scaled']])
```

```python
# Visualize the clusters
plt.figure(figsize=(12, 6))
for i in range(num_clusters):
    cluster_data = data_for_clustering[data_for_clustering['Cluster'] == i]
    plt.scatter(cluster_data['Start Date'], cluster_data['Canada'], label=f'Cluster {i + 1}')
    plt.title('K-Means Clustering of Chicken Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Canada Prices')
plt.legend()
plt.show()
```

```python
# Calculate the difference between consecutive prices
combined_data_eggs['Price Change'] = combined_data_eggs['Canada'].diff()

# Calculate the percentage change between consecutive prices
combined_data_eggs['Percentage Change'] = combined_data_eggs['Canada'].pct_change() * 100
```
