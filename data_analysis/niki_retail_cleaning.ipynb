```python
import pandas as pd
import numpy as np
```

```python
data_1 = pd.read_csv('retail_Sept2022_Sept2023.csv')
print("There are " + str(len(data_1)) + " observations in the retail prices dataset.")
print("There are " + str(len(data_1.columns)) + " variables in the dataset.")
```

```python
data_1.head()
```

```python
# remove tailing number
data_1['Products'] = data_1['Products'].str[:-2]

# split 'Products' using comma and created 2 new columns
data_1['Product Name'] = data_1['Products'].str.split(', ').str[0]
data_1['Units'] = data_1['Products'].str.split(', ').str[1]

# move new columns 
data_1.insert(0, 'Product Name', data_1.pop('Product Name'))
data_1.insert(1, 'Units', data_1.pop('Units'))

# drop 'Products' column
data_1 = data_1.drop(columns = ['Products'])

# split 'Units' using space and create a new column
data_1['Quantity'] = data_1['Units'].str.split(' ').str[0]
data_1['Units'] = data_1['Units'].str.split(' ').str[1]

# move new column 
data_1.insert(1, 'Quantity', data_1.pop('Quantity'))

# change 'per' to 1 in the 'Quantity' column
data_1.loc[data_1['Quantity'] == 'per', 'Quantity'] = '1'

# convert datatype in 'Quantity' to numeric
#data_1['Quantity'] = pd.to_numeric(data_1['Quantity'])

# create 'Average Cost' column
data_1['Average Cost'] = data_1.iloc[:, 3:].mean(axis=1)
data_1['Average Cost'] = data_1['Average Cost'].apply(lambda x: float("{:.2f}".format(x)))

data_1.head()
```

```python
data_1.dtypes
```

```python
data_1.tail(20)
```
