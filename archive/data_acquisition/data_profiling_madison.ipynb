#### Monthly Average Retail Prices for Food and Related Products

```python
# identify which product(s) correspond with the maximum unit price
monthly_avg_retail_prices[monthly_avg_retail_prices['VALUE']==156.8]
```

```python
# Check for total number of records, compare counts across columns to identify missing data
monthly_avg_retail_prices.count()
```

```python
# investigate how many distinct unit types are present
# split Products column based on comma delimiter
data_split = pd.DataFrame(
    [x.rsplit(', ', 1) for x in monthly_avg_retail_prices.Products.tolist()], 
    columns=['product', 'units']
)
data_split 
```

```python
# count number of unique values in the new 'units' column
data_split['units'].unique()
```

```python
data_split['units'].unique().shape
```

```python
# count how many products will need to be split using a different delimiter
data_split['units'].isnull().sum()
```

#### Monthly Consumer Price Index (not seasonally adjusted)

```python
# import data from csv
monthly_cpi_nsa = pd.read_csv(f'{path}/18100004-eng/monthly_cpi_nsa.csv')
monthly_cpi_nsa.head(10)
```

```python
# Noted that not all items are captured when you download all data; csv file itself cannot handle nested product data
```

```python
monthly_cpi_nsa_5yr = pd.read_csv(f'{path}/18100004-eng/1810000401-noSymbol.csv', skiprows=[10], header=8)
monthly_cpi_nsa_5yr
```

```python
# only keep first 15 rows
monthly_cpi_nsa_5yr_clean = monthly_cpi_nsa_5yr.iloc[0:15]
monthly_cpi_nsa_5yr_clean.set_index('Products and product groups 3 4', inplace=True)
```

```python
monthly_cpi_nsa_5yr_clean.dtypes
```

```python
monthly_cpi_nsa_5yr_clean.describe()
```

```python
monthly_cpi_nsa_5yr_tp = monthly_cpi_nsa_5yr_clean.transpose()
monthly_cpi_nsa_5yr_tp
```

```python
monthly_cpi_nsa_5yr_tp = monthly_cpi_nsa_5yr_tp.apply(pd.to_numeric)
monthly_cpi_nsa_5yr_tp.dtypes
```

```python
monthly_cpi_nsa_5yr_tp.describe()
```

```python
monthly_cpi_nsa_5yr_tp.plot(legend=False)
```
