```python
#Data Visualization
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
for df in [df_Pork_2021, df_Pork_2022, df_Pork_2023]:
    df['Year'] = df['Start Date'].dt.year
```

```python
df_combined_pork = pd.concat([df_Pork_2021, df_Pork_2022, df_Pork_2023], ignore_index=True)
```

```python
df_combined_pork.head()
```

```python
# Visualize individual year trends


#Visualize combined trends
plt.figure(figsize=(12, 7))
sns.lineplot(data=df_combined_pork, x='Start Date', y='Canada (2)', hue='Year')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Pork Price Trend in Canada (2021-2023)')
plt.legend(title='Year')
plt.show()
```
